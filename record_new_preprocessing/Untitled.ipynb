```python
import string
import numpy as np
import matplotlib.pyplot as plt
import os
from prettytable import PrettyTable
```

```python
dirs1 = 'bagging_renew_models_PU/noise_0.8_clean_data_size_50_additional_data_size_2000/'
dirs2 = 'bagging_without_renew_models_PU/noise_0.8_clean_data_size_50_additional_data_size_2000/'
```

```python
benchmark_folder = 'benchmark/noise_0.8_lr_0.0003/'
benchmark_accuracy = np.zeros((5, 50))
for index in range(5):
    file = open(benchmark_folder + 'test' + str(index) + '/file_benchmark.txt')
    line = file.readlines()
    benchmark_accuracy[index] = [float(x) for x in line[3].replace('[', '').replace(']', '').replace('\n', '').split(',')]
benchmark_accuracy = np.mean(benchmark_accuracy, axis=0)
print(benchmark_accuracy.shape)
print(benchmark_accuracy)
    
```

```python
# dirs1

for threshold in [0.8, 0.85, 0.9, 0.95]:
    for criterion in [75, 90]:
        folder = dirs1 + 'bagging_threshold_' + str(threshold) + '_add_criterion_' + str(criterion) + \
                '_minimum_additional_size_50/seed_10/'
        precision_file = open(folder + 'precision_file.txt')
        
        line = precision_file.readlines()
        precision = np.zeros((10, 10))
        number = np.zeros((10, 10))
        for i in range(10):
            precision[:, i] = [float(x) for x in line[i*2].replace('[', '').replace(']', '').replace('\n', '').split(',')]
            number[:, i] = [int(x) for x in line[i*2+1].replace('[', '').replace(']', '').replace('\n', '').split(',')]
            
        print(np.around(precision, decimals=3))
        print(number)
        print(folder)
        lambda_list = [0.3, 0.4, 0.5, 0.6, 0.7, 0.8]
        student_accuracy = np.zeros((6, 5, 50))
        teacher_accuracy = np.zeros((5, 50))
        for lambda_i in range(6):
            for index in range(5):
                student_file = open(folder + 'file_student_' + str(index) + '.txt')
                teacher_file = open(folder + 'file_teacher_' + str(index) + '.txt')

                line = student_file.readlines()
                student_accuracy[lambda_i, index, :] = [float(x) for x in line[lambda_i*4+3].replace('[', '').replace(']', '').replace('\n', '').split(',')]
                
                line = teacher_file.readlines()
                accuracy = [float(x) for x in line[3].replace('[', '').replace(']', '').replace('\n', '').split(',')]
                if len(accuracy) < 50:
                    accuracy = accuracy + [accuracy[-1]]*(50-len(accuracy))
                teacher_accuracy[index, :] = accuracy
        student_accuracy = np.mean(student_accuracy, axis=1)
        teacher_accuracy = np.mean(teacher_accuracy, axis=0)
        
        plt.figure(1) # 1
        plt.figure(figsize=(10,6))
        plt.plot(np.arange(50)+1, benchmark_accuracy, '--', label='benchmark')
        plt.plot(np.arange(50)+1, teacher_accuracy, ':', label='teacher')
        for lambda_i in range(6):
            plt.plot(np.arange(50)+1, student_accuracy[lambda_i], label='student, $\lambda$=%.1f'%lambda_list[lambda_i])
        
        plt.legend()
        plt.grid()
        plt.ylim(0,1)
        plt.xlabel('epoch')
        plt.ylabel('accuracy')
        plt.title('confident threshold %.2f, adding criterion %d'%(threshold, criterion))

        dirs = 'figure/' + dirs1
        if not os.path.exists(dirs):
            os.makedirs(dirs)
        plt.savefig(dirs + 'bagging_threshold_' + str(threshold) + '_add_criterion_' + str(criterion) + '_minimum_additional_size_50.png')
        plt.show()
        
        table = PrettyTable(['line', 'Maximum', 'Last'])
        table.add_row(['benchmark', np.round(np.max(benchmark_accuracy), decimals=3), np.round(benchmark_accuracy[-1], decimals=3)])
        table.add_row(['teacher', np.round(np.max(teacher_accuracy), decimals=3), np.round(teacher_accuracy[-1], decimals=3)])
        for i in range(6):
            table.add_row(['student, ='+str(lambda_list[i]), np.round(np.max(student_accuracy[i]), decimals=3), np.round(student_accuracy[i, -1], decimals=3)])
        print(table)
```
