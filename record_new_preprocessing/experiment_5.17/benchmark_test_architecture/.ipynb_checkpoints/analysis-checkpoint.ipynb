```python
import string
import numpy as np
import matplotlib.pyplot as plt
import os
from prettytable import PrettyTable
```

```python
dirs = 'bagging_without_renew_models_PU/noise_0.8_clean_data_size_50_additional_data_size_2000/'
```

```python
# dirs1

for threshold in [0.85]:
    for criterion in [85, 90]:
        folder = dirs + 'bagging_threshold_' + str(threshold) + '_add_criterion_' + str(criterion) + \
                '_minimum_additional_size_50/seed_10/'
        precision_file = open(folder + 'precision_file.txt')
        
        line = precision_file.readlines()
        precision = np.zeros((10, 10))
        number = np.zeros((10, 10))
        for i in range(10):
            precision[:, i] = [float(x) for x in line[i*2].replace('[', '').replace(']', '').replace('\n', '').split(',')]
            number[:, i] = [int(x) for x in line[i*2+1].replace('[', '').replace(']', '').replace('\n', '').split(',')]
            
        print(np.around(precision, decimals=3))
        print(number)
        print(folder)
        lambda_list = [0.8, 0.85, 0.9, 0.95]
        student_accuracy = np.zeros((4, 3, 50))
        for lambda_i in range(4):
            for index in range(3):
                student_file = open(folder + 'bagging_teacher_file_student_' + str(index) + '.txt')
                
                line = student_file.readlines()
                student_accuracy[lambda_i, index, :] = [float(x) for x in line[lambda_i*4+3].replace('[', '').replace(']', '').replace('\n', '').split(',')]
                
        student_accuracy = np.mean(student_accuracy, axis=1)
        
        plt.figure(1) # 寤哄捐〃1
        plt.figure(figsize=(10,6))
       
        for lambda_i in range(4):
            plt.plot(np.arange(50)+1, student_accuracy[lambda_i], label='student, $\lambda$=%.1f'%lambda_list[lambda_i])
        
        plt.legend()
        plt.grid()
        plt.ylim(0,1)
        plt.xlabel('epoch')
        plt.ylabel('accuracy')
        plt.title('confident threshold %.2f, adding criterion %d'%(threshold, criterion))

        dirs1 = 'figure/' + dirs
        if not os.path.exists(dirs1):
            os.makedirs(dirs1)
        plt.savefig(dirs1 + 'bagging_threshold_' + str(threshold) + '_add_criterion_' + str(criterion) + '_minimum_additional_size_50.png')
        plt.show()
        
        table = PrettyTable(['line', 'Maximum', 'Last'])
        for i in range(4):
            table.add_row(['student, 位='+str(lambda_list[i]), np.round(np.max(student_accuracy[i]), decimals=3), np.round(student_accuracy[i, -1], decimals=3)])
        print(table)
```

```python
# compare teacher model with & without bagging
# dirs1

for threshold in [0.8, 0.85, 0.9, 0.95]:
    for criterion in [75, 90]:
        folder = dirs1 + 'bagging_threshold_' + str(threshold) + '_add_criterion_' + str(criterion) + \
                '_minimum_additional_size_50/seed_10/'
        
        print(folder)
        lambda_list = [0.3, 0.4, 0.5, 0.6, 0.7, 0.8]
        student_accuracy = np.zeros((6, 5, 50))
        student_accuracy_with_bagging_teacher = np.zeros((6, 5, 50))
        teacher_accuracy = np.zeros((5, 50))
        for lambda_i in range(6):
            for index in range(5):
                student_file = open(folder + 'file_student_' + str(index) + '.txt')
                student_file_with_bagging_teacher = open(folder + 'bagging_teacher_file_student_' + str(index) + '.txt')
                teacher_file = open(folder + 'file_teacher_' + str(index) + '.txt')

                line = student_file.readlines()
                student_accuracy[lambda_i, index, :] = [float(x) for x in line[lambda_i*4+3].replace('[', '').replace(']', '').replace('\n', '').split(',')]
                
                line = student_file_with_bagging_teacher.readlines()
                student_accuracy_with_bagging_teacher[lambda_i, index, :] = [float(x) for x in line[lambda_i*4+3].replace('[', '').replace(']', '').replace('\n', '').split(',')]
                
                line = teacher_file.readlines()
                accuracy = [float(x) for x in line[3].replace('[', '').replace(']', '').replace('\n', '').split(',')]
                if len(accuracy) < 50:
                    accuracy = accuracy + [accuracy[-1]]*(50-len(accuracy))
                teacher_accuracy[index, :] = accuracy
        student_accuracy = np.mean(student_accuracy, axis=1)
        student_accuracy_with_bagging_teacher = np.mean(student_accuracy_with_bagging_teacher, axis=1)
        teacher_accuracy = np.mean(teacher_accuracy, axis=0)
        
        table = PrettyTable(['line', 'Maximum', 'Last', 'Maximum (with bagging)', 'Last (with bagging)'])
        table.add_row(['benchmark', np.round(np.max(benchmark_accuracy), decimals=3), np.round(benchmark_accuracy[-1], decimals=3), '/', '/'])
        table.add_row(['teacher', np.round(np.max(teacher_accuracy), decimals=3), np.round(teacher_accuracy[-1], decimals=3), '/', '/'])
        for i in range(6):
            table.add_row(['student, 位='+str(lambda_list[i]), np.round(np.max(student_accuracy[i]), decimals=3), np.round(student_accuracy[i, -1], decimals=3), \
                           np.round(np.max(student_accuracy_with_bagging_teacher[i]), decimals=3), np.round(student_accuracy_with_bagging_teacher[i, -1], decimals=3)])
        print(table)
```
